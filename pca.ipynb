```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import fetch_openml
```

```python
# Load MNIST dataset
mnist = fetch_openml('mnist_784', version=1)
data, labels = mnist.data, mnist.target
```

```python
# Standardize data
standardized_data = StandardScaler().fit_transform(data)
```

```python
# Reduce to 2D using PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(standardized_data[:1000, :])
```

```python
# Convert to DataFrame
pca_df = pd.DataFrame(data=np.vstack((pca_data.T, labels[:1000])).T, columns=["Dim_1", "Dim_2", "label"])
```

```python
# Plot
sns.scatterplot(data=pca_df, x="Dim_1", y="Dim_2", hue=pca_df["label"], palette="bright")
plt.title("PCA on MNIST")
plt.show()
```
